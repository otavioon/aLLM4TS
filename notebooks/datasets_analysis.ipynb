```python
from data_provider.data_loader import UEAloader, DAGHAR
from pathlib import Path
import numpy as np
import umap


output_data_path = Path("./output_processed/")
output_data_path.mkdir(parents=True, exist_ok=True)
```

```python
root_data_path = Path("./dataset/")
train_datasets = {}
test_datasets = {}

for dataset in [
    "EthanolConcentration",
    "Handwriting",
    "Heartbeat",
    # "JapaneseVowels",     # Does not have the same shape for all samples in train dataset
    "PEMS-SF",
    "SelfRegulationSCP1",
    "SelfRegulationSCP2",
    # "SpokenArabicDigits", # Does not have the same shape for all samples in train dataset
    "UWaveGestureLibrary",
    "FaceDetection",
]:
    train_output_filename = output_data_path / f"{dataset}_train.npz"
    test_output_filename = output_data_path / f"{dataset}_test.npz"
    
    if train_output_filename.exists() and test_output_filename.exists():
        print(f"Dataset {dataset} already processed")
        continue
    
    print(f"Processing dataset {dataset}....")
    
    train_dataset = UEAloader(root_path=str(root_data_path / dataset), flag="TRAIN")
    xs, ys = [], []
    for i in range(len(train_dataset)):
        x, y = train_dataset[i]
        xs.append(x)
        ys.append(y)
    xs = np.stack(xs)
    ys = np.stack(ys)
    print(f"Dataset (TRAIN): {dataset}. X={xs.shape}, Y={ys.shape}")
    np.savez(train_output_filename, x=xs, y=ys)
    print(f"Saved to {train_output_filename}")
    
    test_dataset = UEAloader(root_path=str(root_data_path / dataset), flag="TEST")
    xs, ys = [], []
    for i in range(len(test_dataset)):
        x, y = test_dataset[i]
        xs.append(x)
        ys.append(y)
    xs = np.stack(xs)
    ys = np.stack(ys)
    print(f"Dataset (TEST): {dataset}. X={xs.shape}, Y={ys.shape}")
    np.savez(test_output_filename, x=xs, y=ys)
    print(f"Saved to {test_output_filename}")
    print("\n")
    
```

```python
import numpy as np
import umap
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def plot_data_allm4ts(X_train, embed_epoch_first, embed_epoch_last, Y_train, title):
    # Flatten the input data
    X_train = X_train.reshape(X_train.shape[0], -1)
    Y_train = Y_train.reshape(-1)
    embed_epoch_first = embed_epoch_first.reshape(embed_epoch_first.shape[0], -1)
    embed_epoch_last = embed_epoch_last.reshape(embed_epoch_last.shape[0], -1)
    
    # Apply UMAP
    print(f"Applying UMAP to X_train...")
    X_train_embed = umap.UMAP(random_state=42, n_components=2).fit_transform(X_train)
    print(f"Applying UMAP to embed_epoch_first...")
    embed_epoch_first_embed = umap.UMAP(random_state=42, n_components=2).fit_transform(embed_epoch_first)
    print(f"Applying UMAP to embed_epoch_last...")
    embed_epoch_last_embed = umap.UMAP(random_state=42, n_components=2).fit_transform(embed_epoch_last)
    
    # Get unique classes and define discrete colormap
    unique_classes = np.unique(Y_train)
    cmap = plt.get_cmap("tab20", len(unique_classes))  # Use 'tab10' or 'Set1' for distinct colors
    norm = mcolors.BoundaryNorm(boundaries=np.arange(len(unique_classes) + 1) - 0.5, ncolors=len(unique_classes))

    # Create subplots
    fig, ax = plt.subplots(1, 3, figsize=(18, 6), constrained_layout=True)
    
    # Scatter plot for X_train
    sc1 = ax[0].scatter(X_train_embed[:, 0], X_train_embed[:, 1], c=Y_train, cmap=cmap, norm=norm, alpha=0.7)
    ax[0].set_title("UMAP Projection of X_train")
    ax[0].set_xlabel("UMAP Dimension 1")
    ax[0].set_ylabel("UMAP Dimension 2")
    cbar1 = fig.colorbar(sc1, ax=ax[0], ticks=unique_classes)
    cbar1.set_label("Classes")
    cbar1.set_ticks(unique_classes)
    cbar1.set_ticklabels(unique_classes)

    # Scatter plot for embed_epoch_first
    sc2 = ax[1].scatter(embed_epoch_first_embed[:, 0], embed_epoch_first_embed[:, 1], c=Y_train, cmap=cmap, norm=norm, alpha=0.7)
    ax[1].set_title("UMAP Projection of aLLM4TS-A (first epoch)")
    ax[1].set_xlabel("UMAP Dimension 1")
    ax[1].set_ylabel("UMAP Dimension 2")
    cbar2 = fig.colorbar(sc2, ax=ax[1], ticks=unique_classes)
    cbar2.set_label("Classes")
    cbar2.set_ticks(unique_classes)
    cbar2.set_ticklabels(unique_classes)

    # Scatter plot for embed_epoch_last
    sc3 = ax[2].scatter(embed_epoch_last_embed[:, 0], embed_epoch_last_embed[:, 1], c=Y_train, cmap=cmap, norm=norm, alpha=0.7)
    ax[2].set_title("UMAP Projection of aLLM4TS-A (last epoch)")
    ax[2].set_xlabel("UMAP Dimension 1")
    ax[2].set_ylabel("UMAP Dimension 2")
    cbar3 = fig.colorbar(sc3, ax=ax[2], ticks=unique_classes)
    cbar3.set_label("Classes")
    cbar3.set_ticks(unique_classes)
    cbar3.set_ticklabels(unique_classes)

    # Set overall title
    plt.suptitle(title, fontsize=16)
    
    # Show plot
    plt.show()
    
import numpy as np

import numpy as np
import pandas as pd

def dataset_statistics(train_X, train_Y):
    # Shape Information
    num_samples, time_steps, num_features = train_X.shape
    
    # Class Distribution in Train Y
    unique_classes, class_counts = np.unique(train_Y, return_counts=True)
    class_distribution = {f"Class {cls}": count for cls, count in zip(unique_classes, class_counts)}
    
    # Overall Statistics for Train X
    overall_mean = np.mean(train_X)
    overall_std = np.std(train_X)
    overall_min = np.min(train_X)
    overall_max = np.max(train_X)

    # Create a DataFrame for dataset summary
    stats_data = {
        "Metric": ["Number of Samples", "Time Steps per Sample", "Number of Features", 
                   "Overall Mean", "Overall Std Dev", "Overall Min", "Overall Max"],
        "Value": [num_samples, time_steps, num_features, overall_mean, overall_std, overall_min, overall_max]
    }
    
    df_stats = pd.DataFrame(stats_data)
    
    # Convert class distribution to a DataFrame
    df_classes = pd.DataFrame(list(class_distribution.items()), columns=["Class", "Sample Count"])
    
    # Display tables
    print("\n### Dataset Overview ###")
    print(df_stats.to_markdown(index=False))

    print("\n### Class Distribution ###")
    print(df_classes.to_markdown(index=False))

# Example usage:
# dataset_statistics(train_X, train_Y)

```
